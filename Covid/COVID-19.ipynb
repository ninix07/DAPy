##  Remove all the records where the Confirmed Cases is Less Than 10.

```python
df=df[~(df.Confirmed<10)]
```

```python
df
```

## In which Region, maximum number of Confirmed cases were recorded ?

```python
df
```

```python
RegionData=df.groupby('Region')
```

```python
RegionData.Confirmed.sum().sort_values(ascending=False)
```

For minimum cases, 

```python
RegionData.Confirmed.sum().sort_values(ascending=True)
```
